```python
# create a range for x
x = list(range(10))
# use the formual y = x^2 - x + 2 to get the values for y
y = [x**2 - x + 2 for x in x]

# Use plt.plot to create a basic line graph
plt.plot(x, y)
# use plt.annotate to lable the origin point
plt.annotate('Origin', fontsize = 14, xy=(0,0), xytext=(1, 25), arrowprops={'facecolor' : 'blue'})
# use plt.show to show the result
plt.show()
```

# 2

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
- y = $\sqrt{x}$
- y = $x^3$
- y = $2^x$

```python
import math
print('Success!!')
```

```python
x = list(range(257))
y = [math.sqrt(x) for x in x]

plt.plot(x, y)
plt.title("y = $\sqrt{x}$ \n where x in range 257" , fontsize = 16)
plt.show()
```

```python
x = list(range(6))
y = [math.pow(n, 3) for n in x]


plt.plot(x, y)
plt.title("y = $x^3$ \n where x in range 6", fontsize = 16)
plt.show()
```

```python
x = list(range(6))
y = [2**n for n in x]

plt.plot(x, y)
plt.title("y = $2^x$ \n where x in range 6", fontsize = 18)
plt.show()
```

```python
plt.figure(figsize = (20,20))
plt.suptitle("1 for all and all 4 one", fontsize = 24)

plt.subplot(221)
x = list(range(10))
y = [x**2 - x + 2 for x in x]
plt.plot(x, y)
plt.title("y = $x^2$ - x + 2 for x in x \n where x in range 10", fontsize = 16)
plt.annotate('Origin', fontsize = 14, xy=(0,0), xytext=(1, 25), arrowprops={'facecolor' : 'blue'})

plt.subplot(222)
x = list(range(257))
y = [math.sqrt(x) for x in x]

plt.plot(x, y)
plt.title("y = $\sqrt{x}$ \n where x in range 257" , fontsize = 16)

plt.subplot(223)
x = list(range(6))
y = [math.pow(n, 3) for n in x]

plt.plot(x, y)
plt.title("y = $x^3$ \n where x in range 6", fontsize = 16)

plt.subplot(224)
x = list(range(6))
y = [2**n for n in x]
plt.plot(x, y)
plt.title("y = $2^x$ \n where x in range 6", fontsize = 18)
plt.show()
```

```python
x1 = list(range(10))
y1 = [x**2 - x + 2 for x in x1]

x2 = list(range(10))
y2 = [math.sqrt(x) for x in x2]

x3 = list(range(10))
y3 = [math.pow(n, 3) for n in x3]

x4 = list(range(10))
y4 = [2**n for n in x4]

plt.figure(figsize = (10,5))
plt.scatter(x1, y1, c="red", label="y = $\sqrt{x}$")
plt.scatter(x2, y2, c ="green", label="y = $\sqrt{x}$")
plt.scatter(x3, y4, c="blue", label="y = $x^3$")
plt.scatter(x4, y4, c="orange", label="y = $2^x$", alpha=.7)
plt.title("y = $x^2$ - x + 2 for x in x \n where x in range 10", fontsize = 16)
plt.figure
plt.legend(loc='upper left', fontsize = "14")
#plt.annotate('Origin', fontsize = 14, xy=(0,0), xytext=(0, 10), arrowprops={'facecolor' : 'blue'})
plt.show() 
```
