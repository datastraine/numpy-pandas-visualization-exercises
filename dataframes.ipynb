```python
# Load built in data from pydataset

from pydataset import data

data
```

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

stugrades = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

stugrades
```

```python
stugrades['passing_english'] = stugrades.english >= 63
stugrades
```

```python
stugrades.sort_values(by='passing_english')

# Duplicate values are sorted in index order
```

```python
stugrades.sort_values(by='name').sort_values(by='passing_english')
```

```python
stugrades.sort_values(by='english').sort_values(by='passing_english')
```

```python
stugrades['over_all_grade'] = (stugrades['math'] + stugrades['english'] + stugrades['reading'])\
                              /len(['math', 'english', 'reading'])

stugrades
```

```python
# Rename the cty column to city.
# Rename the hwy column to highway.

mpg.rename(columns={'cty': 'city'}, inplace = True)
mpg.rename(columns={'hwy': 'highway'}, inplace = True)
mpg
```

```python
# Do any cars have better city mileage than highway mileage? - No

mpg[mpg['city'] > mpg['highway']]
```

```python
# create a mileage_difference variable where it is the difference between the highway and city driving

mpg['mileage_difference'] = mpg['highway'] - mpg['city']
mpg
```

```python
# create a bool value of the milage difference column where...
# the miliage difference is equal to the max milage difference and return the corresponding cars

# pull all values in the data frame where the milage difference is equal to the max miliage difference
mpg[mpg['mileage_difference'] == mpg['mileage_difference'].max()]
```

```python
# Which compact class car has the lowest highway mileage? 
#mpg.sort_values(by='highway').head(1)

# Where the class is compact, sort the values by highway milage in descending order and pull the lowest value
mpg[mpg["class"] == 'compact'].sort_values(by='highway', ascending=False).tail(1)
```

```python
#The best?

# Where the class is compact, sort the values by highway milage in descending order and pull the top value
mpg[mpg["class"] == 'compact'].sort_values(by='highway', ascending = False).head(1)
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg['highway'] + mpg['city'])/len(['highway', 'city'])
mpg
```

```python
# Which dodge car has the best average mileage? 

# Where the manufacturer is dodge, sort the values by the average_mileage in descending order and pull the top value
mpg[mpg['manufacturer'] == 'dodge'].sort_values(by='average_mileage', ascending = False).head(1)
```

```python
# The worst?

mpg[mpg['manufacturer'] == 'dodge'].sort_values(by='average_mileage', ascending=False).tail(1)
```

```python
# What is the the weight of the fastest animal?

# Assign the max speed to a variable
fastest = mammals['speed'].max()

# inside the data frame pull the weight column where the speed is equal to the fastest
mammals['weight'][mammals['speed'] == fastest]
```

```python
# What is the overal percentage of specials?

# Get a percentage of the True and false statements 
percent_speicals = mammals['specials'].value_counts(True) * 100

# Only pull the value where the statement is true
percent_speicals[True]
```

```python
# How many animals are hoppers that are above the median speed? What percentage is this?

# Find the median speed for all mammals
med_speed = mammals['speed'].median()

# For all hoppers, get a count of all true or false values where hoppers speed is or is not > med_speed
hoppers = mammals['hoppers'][mammals["speed"] >med_speed].value_counts()

# Pull the number of hoppers where the value is True from the previous step  
hoppers_above_med = hoppers[True]

#Divide the hoppers_above_med by the total number of rows in the DF
hoppers_above_med/len(mammals) * 100
```
